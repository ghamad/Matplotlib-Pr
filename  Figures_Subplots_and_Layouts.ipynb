```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(facecolor=(1,0,0,0.1))
plt.show()
```

```python
fig = plt.figure(figsize=plt.figaspect(2.), facecolor=(1,0,0.6,0.1))
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.set(xlim=[0.5, 4.5], ylim=[-2, 8], title="An example axes",
      xlabel='X-axis', ylabel='Y-axis')
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.set_xlim([0.5, 4.5])
ax.set_ylim([-2, 8])
ax.set_title('A Different Example Axes Title')
ax.set_ylabel('Y-Axis (changed)', size=25)
ax.set_xlabel('X-Axis (changed)')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,2,3,4],[10,20,30,40], color='blue', linewidth=3)
ax.scatter([0.3,1.5,3.56,6.45], [11,25,7,20], c=[1,3,4,2], marker='^')
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,30,40], color='blue', linewidth=3)
plt.scatter([0.3,1.5,3.56,6.45], [11,25,7,20], c=[1,3,4,2], marker='^')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2)
axes[0,1].set(title='Upper Right')
axes[1,0].set(title='Lower Left')
axes[1,1].set(title='Lower Right')

# To iterate over all items in a multidimensional numpy array, use the `flat` attribute
for ax in axes.flat:
    # Remove all xticks and yticks...
    ax.set(xticks=[5], yticks=[5])
    
plt.show()
```
