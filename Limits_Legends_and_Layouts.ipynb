```python
from __future__ import print_function

import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=plt.figaspect(0.5))

ax1.plot([-10, -5, 0, 5, 10, 15], [-1.2, 2, 3.5, -0.3, -4, 1])
ax2.scatter([-10, -5, 0, 5, 10, 15], [-1.2, 2, 3.5, -0.3, -4, 1])

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=plt.figaspect(0.5))

ax1.plot([-10, -5, 0, 5, 10, 15], [-1.2, 2, 3.5, -0.3, -4, 1])
ax2.scatter([-10, -5, 0, 5, 10, 15], [-1.2, 2, 3.5, -0.3, -4, 1])

#amrgins value are given from the range of 0-1 as in fraction.
ax1.margins(x= 0.0, y=0.1)
ax2.margins(0.5)
ax1.axis('equal')
ax2.axis('tight')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=3)

for ax in axes:
    ax.plot([-10, -5, 0, 5, 10, 15], [-1.2, 2, 3.5, -0.3, -4, 1])
    
axes[0].set_title('Normal Autoscaling', y=0.7, x=0.8)

axes[1].set_title('ax.axis("tight")', y=0.7, x=0.8)
axes[1].axis("tight")

axes[2].set_title('ax.axis("equal")', y=0.7, x=0.8)
axes[2].axis("equal")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('classic')

t = np.linspace(0, 2 * np.pi, 150)
x1, y1 = np.cos(t), np.sin(t)
x2, y2 = 2 * x1, 2 * y1

colors = ['darkred', 'darkgreen']

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x1, y1, label='Inner', color='darkred', lw=3)
ax.plot(x2, y2, label='Outer', color='darkgreen', lw=3)
ax.margins(0.1)
ax.axis('equal')
ax.legend()
plt.show()
```

```python
def example_plot(ax):
    ax.plot(1, 2)
    ax.set_xlabel('x-axis', fontsize=16)
    ax.set_ylabel('y-axis', fontsize=8) 
    ax.set_title('Title', fontsize=24)   
    
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)

fig.tight_layout()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)
ax1.plot([1, 2, 3, 4], [1, 2, 3, 4])
ax2.plot([3, 4, 5, 6], [6, 5, 4, 3])
plt.show()
```

```python
fig, ax1 = plt.subplots(1, 1)
ax1.plot([1, 2, 3, 4], [1, 2, 3, 4])
ax2 = ax1.twinx()
ax2.scatter([1, 2, 3, 4], [60, 50, 40, 30])
ax1.set(xlabel='X', ylabel='First scale')
ax2.set(ylabel='Other scale')
plt.show()
```
